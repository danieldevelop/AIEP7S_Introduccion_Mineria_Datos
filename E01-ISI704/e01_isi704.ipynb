# Caso 3

Cree un nuevo conjunto de datos, en los cuales deba “hacer limpieza de datos”, tomando como ejemplo las cabeceras del csv creado, con información nueva, con la cual usted deba reemplazar información y rellenar con valores los registros faltantes, en la casilla apellido con la frase “No hay registro”

```python
estudiante = {
    "nombre": ["Soledad", "Marta", "Simon", "Maria", "Pedro"],
    "apellido": ["No hay registro", "Perez", "No hay registro", "Gomez", "Perez"],
    "edades": [15, 20, np.nan, 38, 25],
    "semestre": [3, 5, 8, 1, np.nan]
}

dfEstudiantes = pd.DataFrame(estudiante)
dfEstudiantes
```

```python
# Remplazar valores faltantes
updateEstudiante = dfEstudiantes.replace(
    {
        "apellido": {
            "No hay registro": "Desconocido"
        },
    }
)
updateEstudiante
```

```python
# Rellenar valores faltantes
fillEstudiante = dfEstudiantes.fillna(
    {
        "edades": dfEstudiantes["edades"].mean(),
        "semestre": dfEstudiantes["semestre"].mode()[0]
    }
)
fillEstudiante
```
